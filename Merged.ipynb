```python
crime_df = pd.read_csv('Output/crime_data.csv', encoding='UTF-8')
crime_df
```

```python
weather_df = pd.read_csv('Output/weather_data.csv', encoding='UTF-8')
weather_df
```

```python
income_df = pd.read_csv('Output/income_data.csv', encoding='UTF-8')
income_df
```

```python
income_df1 = income_df.rename(columns={"City": "GEO"})
income_df1
```

```python
employment_shelter_df = pd.read_csv('Output/shelter_and_emp_df.csv', encoding='UTF-8')
employment_shelter_df
```

```python
merged_df1 = crime_df.merge(weather_df, on='City_ID2', how='inner')
merged_df1
```

```python
merged_df2 = merged_df1.merge(income_df1, left_index=True, right_index=True)
merged_df2
```

```python
columns_to_keep = ['GEO', 'City_y', 'Crime Severity Index Value in 2021','Median Income', 'Wind Speed', 'Max Temp']
merged_df3 = merged_df2[columns_to_keep]
merged_df3
```

```python
merged_df4 = merged_df3.merge(employment_shelter_df, left_index=True, right_index=True)
merged_df4
```

```python
columns_to_keep1 = ['GEO', 'City_y', 'Crime Severity Index Value in 2021','Median Income', 'Wind Speed', 'Max Temp', 'Median monthly shelter costs for owned dwellings ($)', 'Average monthly shelter costs for owned dwellings ($)', 'Median monthly shelter costs for rented dwellings ($)', 'Average monthly shelter costs for rented dwellings ($)', 'Employment Rate (%)']
merged_df5 = merged_df4[columns_to_keep1]
merged_df5
```

```python
#merged_df5["Median Income/Average Monthly Rent Cost"] = merged_df5["Median Income"] / merged_df5["Average monthly shelter costs for rented dwellings ($)"]
merged_df5['Median Income/Average Monthly Rent Cost'] = merged_df5.apply(lambda x: x['Median Income'] / x['Average monthly shelter costs for rented dwellings ($)'], axis=1)
merged_df5
```

```python
merged_df5['Median Income/Average Monthly Shelter Cost for Owned Dwellings'] = merged_df5.apply(lambda x: x['Median Income'] / x['Average monthly shelter costs for owned dwellings ($)'], axis=1)
merged_df5
```

```python
merged_df6 = merged_df5.rename(columns={"GEO": "Long Name", "City_y": "Name"})
merged_df6
```

```python
# Export the merged data into a csv

merged_df6.to_csv("Output/merged_data.csv")
```
