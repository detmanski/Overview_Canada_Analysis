```python
import pandas as pd
```

```python
# load csv data for crime data
crime_df = pd.read_csv('Output/crime_data.csv', index_col=False, encoding='UTF-8')
crime_df
```

```python
# load csv data for weather data
weather_df = pd.read_csv('Output/weather_data.csv', index_col=False, encoding='UTF-8')
weather_df
```

```python
# load csv data for income data
income_df = pd.read_csv('Output/income_data.csv', index_col=False, encoding='UTF-8')
income_df
```

```python
income_df.rename(columns={"Rank": "Income Ranking"}, inplace=True)
```

```python
# load csv data for employment and shelter data
employment_shelter_df = pd.read_csv('Output/shelter_and_emp_df.csv', index_col=False, encoding='UTF-8')
employment_shelter_df
```

```python
# reduce columns
columns_to_keep_employment_shelter = ['City.1', 'Median monthly shelter costs for owned dwellings ($)', 'Average monthly shelter costs for owned dwellings ($)', 'Median monthly shelter costs for rented dwellings ($)', 'Average monthly shelter costs for rented dwellings ($)', 'Employment Rate (%)']
employment_shelter_df1 = employment_shelter_df[columns_to_keep_employment_shelter]
employment_shelter_df1
```

```python
employment_shelter_df1.rename(columns={"City.1": "GEO_NAME"}, inplace=True)
```

```python
# load csv data for culture data
condensed_culture_df = pd.read_csv('Output/condensed_culture_df.csv', index_col=False, encoding='UTF-8')
condensed_culture_df
```

```python
# bring in lookup to convert city name to GEO_NAME
cities_lookup_df = pd.read_csv('Resources/cities_lookup.csv')
lookup_columns_to_keep = ['Name', 'GEO_NAME']
reduced_lookup_df = cities_lookup_df[lookup_columns_to_keep]
reduced_lookup_df
```

```python
culture_df = condensed_culture_df.merge(reduced_lookup_df, on='Name', how='outer')
culture_df
```

```python
columns_to_keep_culture = ['GEO_NAME', 'Name', 'Participation in Cultural Activities (%)']
culture_df1 = culture_df[columns_to_keep_culture]
culture_df1
```

```python
merged_df1 = crime_df.merge(weather_df, on='GEO_NAME', how='inner')
merged_df1
```

```python
merged_df2 = merged_df1.merge(income_df, on='GEO_NAME', how='outer')
merged_df2
```

```python
merged_df3 = merged_df2.merge(employment_shelter_df1, on='GEO_NAME', how='outer')
merged_df3
```

```python
merged_df4 = merged_df3.merge(culture_df1, on='GEO_NAME', how='outer')
merged_df4
```

```python
merged_df4['Median Income/Average Monthly Rent Cost'] = merged_df4.apply(lambda x: x['Median Income'] / x['Average monthly shelter costs for rented dwellings ($)'], axis=1)
merged_df4
```

```python
merged_df4['Median Income/Average Monthly Shelter Cost for Owned Dwellings'] = merged_df4.apply(lambda x: x['Median Income'] / x['Average monthly shelter costs for owned dwellings ($)'], axis=1)
merged_df4
```

```python
merged_df4.rename(columns={"GEO_NAME": "Long Name"}, inplace=True)
```

```python
# Export the merged data into a csv

merged_df4.to_csv("Output/merged_data.csv", index=False)
```
