```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
merged_df = pd.read_csv('../Output/merged_data.csv', index_col=False, encoding='UTF-8')
merged_df
```

```python
for col in merged_df.columns:
    print(col)
```

```python
%%capture --no-display

# Configure the map plot
map_plot = merged_df.hvplot.points(
    "Lng",
    "Lat",
    geo = True,
    tiles = "CartoLight",
    frame_width = 700,
    frame_height = 500,
    size = "Employment Rate (%)",
    scale = 1.8,
    color = "Name",
    hover_cols = ['Name', 'Percent of Income Spent on Shelter (Rent)', 'Percent of Income Spent on Shelter (Owned)', 'Employment Rate (%)', 'Crime Severity Index Value in 2021']
)

# Display the map
map_plot
```
