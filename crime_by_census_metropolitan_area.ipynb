```python
df = pd.read_csv('Resources/35100026.csv', encoding='utf-8')
#, encoding='ISO-8859-1'

df.head()
```

```python
columns_to_keep=['REF_DATE','GEO','Statistics','DGUID','VALUE']
reduced_df=df[columns_to_keep]
```

```python
#cut down columns for readability
reduced_df.head()
```

```python
filtered_df = reduced_df[reduced_df['Statistics'] == 'Crime severity index']
```

```python
filtered_df.head(7)
```

```python
filtered_df1 = filtered_df[filtered_df['REF_DATE'] == 2021]
```

```python
filtered_df1
```

```python
filtered_df1[['GEO','Province']] = filtered_df1['GEO'].str.split(',',expand=True)
filtered_df1
```

```python
filtered_df2 = filtered_df1.dropna()
filtered_df2
```

```python
filtered_df2.index.name = 'City_ID2'
filtered_df2
```

```python
cities_lookup_df = pd.read_csv('Resources/cities_lookup.csv')
#, encoding='ISO-8859-1'

cities_lookup_df
```

```python
cities_lookup_df1 = cities_lookup_df.dropna()
cities_lookup_df1
```

```python
columns_to_keep = ['City_ID', 'City', 'City_ID2']
reduced_lookup_df = cities_lookup_df1[columns_to_keep]
reduced_lookup_df
```

```python
cities_lookup_df2 = reduced_lookup_df.astype({'City_ID2':'int'})
cities_lookup_df2
```

```python
merged_df = filtered_df2.merge(cities_lookup_df2, on='City_ID2', how='inner')
merged_df
```

```python
columns_to_keep_crime = ['City_ID2', 'City', 'VALUE']
crime_df = merged_df[columns_to_keep_crime]
crime_df
```

```python
crime_df1 = crime_df.rename(columns={"VALUE": "Crime Severity Index Value in 2021"})
```
