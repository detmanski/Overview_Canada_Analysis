```python
radius= 10000
city = condensed_df["Name"] + "," + condensed_df["Country"]
params= {"categories": "entertainment.culture",
        "apiKey": geoapify_key}
print("Starting Search")
for index, row in condensed_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    params["filter"] = f"circle:{lng},{lat},{radius}"
    params["bias"] = f"proximity:{lng}, {lat}"
    base_url = "https://api.geoapify.com/v2/places"
    name_address = requests.get(base_url, params=params)
    name_address = name_address.json()
   
     try:
        condensed_df.loc[index, "Places"] = name_address
    except (KeyError, IndexError):
            condensed_df.loc[index, 'Places'] = "no places found"
    
    print(f"{condensed_df.loc[index, 'Name']} - nearest place: {condensed_df.loc[index, 'Places']}")
```
