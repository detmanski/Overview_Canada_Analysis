```python
#Quick plot showing only one metric of activities by province
culture_vis_df.plot(kind="bar", x= "Province", y= "Making Music(%)", figsize=(4,4))
plt.xlabel("Making Music")
plt.ylabel("Percent Participation")
plt.show()
plt.tight_layout()
```

```python
#Pulling the condensed df to plot in a way that ties into the rest of the presentation
string = 'Resources/condensed_culture_df.csv'
condensed_df = pd.read_csv(string)
condensed_df
```

```python
#Showing the overall participation by city as a overall vies
condensed_df.plot(kind="bar", x= "Name", y= "Participation in Cultural Activities (%)", figsize=(6,6), legend=False)
plt.xlabel("City")
plt.ylabel("Percent Participation")
plt.title("Overall Participation in Cultural Activities by City")
plt.savefig("cities_overall_participation.png")
plt.show()
plt.tight_layout()
```

```python
#Creating keywords to use in visualizations
making_music = culture_vis_df["Making Music(%)"]
dancing = culture_vis_df["Dancing(%)"]
visual_arts = culture_vis_df["Visual Arts(%)"]
theatre = culture_vis_df["Theatre(%)"]
crafting = culture_vis_df["Crafting(%)"]
writing = culture_vis_df["Writing(%)"]
auto_visual = culture_vis_df["Audio-Visual and Interactive Media(%)"]
other = culture_vis_df["Other Cultural Activities(%)"]
labels = culture_vis_df["Province"]
```

```python
#Stacked bar graph code 
fig,ax = plt.subplots(1, figsize=(8,8))
ax.bar(labels, crafting, width=(0.75), label= "Crafting")
ax.bar(labels, making_music, width=(0.75), label="Music")
ax.bar(labels, visual_arts, width=(0.75), label= "Visual Art")
ax.bar(labels, writing, width=(0.75), label= "Writing")
ax.bar(labels, dancing, width=(0.75), label="Dancing")
ax.bar(labels, other, width=(0.75), label= "Other")
ax.bar(labels, auto_visual, width=(0.75), label="AV")
ax.bar(labels, theatre, width=(0.75), label="Theatre")
ax.set_ylabel("Percent Participation")
ax.set_xlabel("Province")
ax.set_title("Participation In Cultural Activities")
ax.legend(loc='upper left')
pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.9, pos.height])
ax.legend(loc='center right', bbox_to_anchor=(1.25, 0.5))
plt.show()
```
