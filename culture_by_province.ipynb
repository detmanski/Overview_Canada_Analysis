```python
import os
import pandas as pd
import matplotlib as plt
```

```python
bc_string = 'Resources/BC_Cultural_Participation.csv'
bc_df = pd.read_csv(bc_string)
bc_df.head()
```

```python
bc_df.drop(columns=["REF_DATE","DGUID","Age group","Sex","Current employment status", "Perceived health", "VECTOR", "COORDINATE"], inplace=True)
bc_df.drop(columns=["VALUE", "STATUS", "SYMBOL", "TERMINATED", "DECIMALS"], inplace=True)
bc_df
```

```python
bc_df = bc_df.rename(columns= {"GEO": "Province",
                              "UOM": "Type",
                               "UOM_ID": "Count",
                              "SCALAR_FACTOR": "Scale",
                              })
bc_df.drop(columns="SCALAR_ID", inplace=True)
```

```python
bc_df["City"] = "Vancouver"
bc_df
```

```python
bc_df
bc_df.to_csv("Resources/bc_df.csv")
```
